```python
import pandas as pd
data_students = pd.read_csv("C:/Users/lgnavarrete/Documents/FundamentalsPython/clean_students_complete.csv")
```

```python
def Func1(y,x):
    print("Grafica de estudiantes por asignatura con puntaje de: ",x)
    asig_min = data_students[data_students[y] == x]
    gender_scname = asig_min[["gender", "school_name"]]
    dummies_genero = pd.get_dummies(gender_scname["gender"])
    plotbygen = pd.concat([gender_scname,dummies_genero],axis = 1)
    plotgrouped = plotbygen.groupby("school_name").sum()
    return plotgrouped.plot.bar(figsize = (12,7))
```

```python
Func1("reading_score",data_students["reading_score"].min())
```

```python
Func1("reading_score",data_students["reading_score"].max())
```
